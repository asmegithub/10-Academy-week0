```python
# overview of the dataset
data.info()
data.isnull().sum()
```

```python
# Drop columns with all NaN values (# comments)
data = data.dropna(axis=1, how='all')
data.head()
```

```python
# Global Horizontal Irradiance (GHI),Direct Normal Irradiance (DNI), and Diffuse Horizontal Irradiance (DHI) should have all non-negative values!!
# so we should replace -ve values with NaN
data[['GHI', 'DNI', 'DHI']] = data[['GHI', 'DNI', 'DHI']].map(lambda x: x if x >= 0 else None)

# Convert Timestamp to datetime
data['Timestamp'] = pd.to_datetime(data['Timestamp'])

# Verify replacement
negative_values_check = (data[['GHI', 'DNI', 'DHI']] < 0).sum()
print("Negative Values After Replacement:\n", negative_values_check)
```

```python
# Check for negative values
negative_values_final_check = (data[['GHI', 'DNI', 'DHI']] < 0).sum()
print("\nNegative Values Final Check:\n", negative_values_final_check)

# Check for missing values in the entire dataset
missing_values_final_check = data.isnull().sum()
print("\nMissing Values Final Check:\n", missing_values_final_check)
```

```python
# Replace NaN with the mean of each column
for col in ['GHI', 'DNI', 'DHI']:
    data[col].fillna(data[col].mean(),inplace=True)

# Verify if missing values remain
missing_values_check = data[['GHI', 'DNI', 'DHI']].isnull().sum()
print("\n Missing Values After Imputation:\n", missing_values_check)
```
