```python
# Data quality checks

# Check for negative values in all columns
negative_values = (data.select_dtypes(include=['float64', 'int64']) < 0).sum()
print("Negative Values in Numeric Columns:\n", negative_values)

# Check for missing values
missing_values = data.isnull().sum()
print("\n\n Missing Values in All Columns:\n", missing_values)
```

```python
# Drop columns with all NaN values (# comments)
data = data.dropna(axis=1, how='all')
data.head()
```

```python
# Global Horizontal Irradiance (GHI),Direct Normal Irradiance (DNI), and Diffuse Horizontal Irradiance (DHI) should have all non-negative values!!
# so we should replace -ve values with NaN
data[['GHI', 'DNI', 'DHI']] = data[['GHI', 'DNI', 'DHI']].map(lambda x: x if x >= 0 else None)

# Convert Timestamp to datetime
data['Timestamp'] = pd.to_datetime(data['Timestamp'])

# Verify replacement
negative_values_check = (data[['GHI', 'DNI', 'DHI']] < 0).sum()
print("Negative Values After Replacement:\n", negative_values_check)
```

```python
# Replace NaN with the mean of each column
for col in ['GHI', 'DNI', 'DHI']:
    data[col].fillna(data[col].mean())

# Verify if missing values remain
missing_values_check = data[['GHI', 'DNI', 'DHI']].isnull().sum()
print("\n Missing Values After Imputation:\n", missing_values_check)
```

```python

# Check for negative values
negative_values_final_check = (data[['GHI', 'DNI', 'DHI']] < 0).sum()
print("\nNegative Values Final Check:\n", negative_values_final_check)

# Check for missing values in the entire dataset
missing_values_final_check = data.isnull().sum()
print("\nMissing Values Final Check:\n", missing_values_final_check)



```

```python
# Column Distributions: histogram
data[['GHI', 'DNI', 'DHI']].hist(bins=20, figsize=(10, 6))
plt.show()
```

```python
# Correlation heatmap
# 1. Focused Analysis of Solar Metrics
# GHI, DNI, and DHI are the primary metrics used in solar energy studies.
# These metrics have well-defined relationships: GHI = DNI  cos() + DHI (where  is the solar zenith angle).

sns.heatmap(data[['GHI', 'DNI', 'DHI']].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Between Solar Metrics")
plt.show()
```

```python
# Analyze Weather Factors:

# Include Tamb (ambient temperature), RH (relative humidity), and WS (wind speed and identify their influence on solar irradiance.
sns.pairplot(data[['GHI', 'DNI', 'DHI', 'Tamb', 'RH', 'WS']])
plt.show()
```

```python
# Aggregate Data by Time:

data['Timestamp'] = pd.to_datetime(data['Timestamp'])
data['Hour'] = data['Timestamp'].dt.hour
hourly_averages = data.groupby('Hour')[['GHI', 'DNI', 'DHI']].mean()

hourly_averages.plot(figsize=(10, 6))
plt.title("Average Solar Irradiance by Hour")
plt.xlabel("Hour")
plt.ylabel("Irradiance")
plt.show()
```
